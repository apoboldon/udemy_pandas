```python
import pandas as pd
```

```python
df=pd.read_csv('imdb.csv')
df
```

```python
df[(df['year']>2012) & (df['genre']== 'action')] # elegir películas posteriores al año 2012 y de género acción. si ponemos el signo | significa "o". hará que se cumpla una cosa o la otra.
```

### ver si están en la lista las películas star Wars y Blade

```python
dfsel=df[df['name'].isin(['Star Wars', 'Blade', 'Iron Man 3'])]   # con el comando isin() busca las películas que le decimos.
dfsel
```

### ordenar el índice, para que no haya saltos

```python
dfsel.reset_index(drop=True) # así riniciamos el índice desde 0 y al poner drop=True, elimina la columna de index que genera pandas automaticamente.
```
